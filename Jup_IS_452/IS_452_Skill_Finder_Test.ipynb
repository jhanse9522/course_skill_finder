```python
import collections
from typing import Dict, List, Tuple

def stemmed2bow(words: List[str], dictionary: Dict[str, int]) -> List[Tuple[int, int]]:
    word_count = collections.defaultdict(int)
    for word in words:
        if word not in dictionary:
            dictionary[word]= len(dictionary)  # if not in, get base going
        word_count[dictionary[word]] +=1         # increment value of k/v pair each time word found
                   
    return list(word_count.items())             # return list of key/value pairs

string_of_stemmed = ",".join(stemmed)

dictionary = {}

print("Bag of Words representation:", stemmed2bow(string_of_stemmed.split(","), dictionary))

print(type(stemmed2bow(string_of_stemmed.split(","), dictionary)))
                   
```
