*NLTK Tree*


```python
fhand = open("/Users/jamey_hansen/Documents/Pycharm_projects/IS_452/Final_Project/git_452_final/DPS_Stage4", "r")
filein = fhand.read()  # my text string
fhand.close()

lines = filein.split("/n")  # my text as list of lines

import spacy
from nltk import Tree

nlp = spacy.load('en')

parsed_courses = nlp(filein)

def to_nltk_tree(root_node):
    if root_node.n_lefts + root_node.n_rights > 0:
        parsed_child_nodes = [to_nltk_tree(child) for child in root_node.children]
        return Tree(root_node.orth_, parsed_child_nodes)
    else:
        return root_node.orth_
    
for sent in parsed_courses.sents:
    to_nltk_tree(sent.root).pretty_print()
```
